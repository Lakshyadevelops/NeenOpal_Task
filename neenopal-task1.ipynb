```python
data1=pd.read_csv('/kaggle/input/record-matching/data1.csv')
data2=pd.read_csv('/kaggle/input/record-matching/data2.csv')
```

```python
data1
```

```python
both_count=0
data1_entries=data1.shape[0]
data2_entries=data2.shape[0]
sum_qty_only_in_data2=0


#created new column which can act as primary key
data1['key']=data1['Order ID']+data1['Product ID']
data2['key']=data2['Order ID']+data2['Product ID']
dict1={}
dict2={}
#   created dictionary for both data

for index1,row1 in data1.iterrows():
    dict1[row1['key']]=1
for index2,row2 in data2.iterrows():
    dict2[row2['key']]=1 

#   calculated common entries
for index1,row1 in data1.iterrows():
    if(0!=dict2.get(row1['key'],0)):
        both_count+=1
        
#   calculating qty for entries in data2 only
for index2,row2 in data2.iterrows():
    if(0!=dict1.get(row2['key'],0)):
        sum_qty_only_in_data2+=row2['Qty']
        
```
